```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as mt
```

```python
data=pd.read_csv("C:\\Users\\arjun\\Documents\\data analytics\\Netflix Userbase.csv")
```

```python
data.head(10)
```

```python
data.describe()
```

```python
sns.countplot(x='Gender',data=data)
mt.title('gender des')
mt.xlabel('gender')
mt.ylabel('count')
mt.show()
```

```python
sns.countplot(x='Gender', data=data,hue='Gender', color='red')
mt.title('chart')
mt.xlabel('gender')
mt.ylabel('count')
mt.show()
```

```python
subscription_counts= data['Subscription Type'].value_counts()
mt.pie(subscription_counts,labels=subscription_counts.index,autopct='%1.1f%%',startangle=90)
mt.title('subscription Types')
mt.show()
```

```python
sns.boxplot(x='Gender',y='Monthly Revenue',data=data)
mt.title('Monthly Revenue by gender')
mt.xlabel('Gender')
mt.ylabel('Monthly Revenue')
mt.show()
```

```python
corr_matrix = data[['Monthly Revenue','Age']].corr()
sns.heatmap(corr_matrix,annot=True,cmap='crest')
mt.title('Correlation Matrix')
mt.show()
```

```python
sns.countplot(y='Country',data=data)
mt.title('country')
mt.xlabel('country')
mt.ylabel('count')
```
